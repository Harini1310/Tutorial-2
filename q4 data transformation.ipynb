```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
data=pd.read_csv(r"C:\\Users\\Harini\\Desktop\\sem7\\DA\\DA LAB\\StudentsPerformance.csv")
df = pd.DataFrame(data)
min_max=MinMaxScaler()
print("MIM MAX NORMALIZATION ")
print(min_max.fit_transform(df[['mathscore']]))

print(min_max.fit_transform(df[['readingscore']]))

print(min_max.fit_transform(df[['writingscore']]))


print("Z-SCORE NORMALIZATION")

a=np.array(df['mathscore'])
print(stats.zscore(a))

b=np.array(df['readingscore'])
print(stats.zscore(b))

c=np.array(df['writingscore'])
print(stats.zscore(c))

```
