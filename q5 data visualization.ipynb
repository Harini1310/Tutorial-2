```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv(r"C:\Users\Harini\Desktop\sem7\DA\DA LAB\StudentsPerformance.csv")
```

```python
fig=plt.figure() #Plots in matplotlib reside within a figure object, use plt.figure to create new figure
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.hist(df['mathscore'],bins = 7) # Here you can play with number of bins

plt.title('MATHS SCORE DISTRIBUTION')
plt.xlabel('MATHS SCORE')
plt.ylabel('#STUDENTS')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
#Variable
ax.boxplot(df['writingscore'])
plt.show()
```

```python
var = df.groupby('gender').mathscore.sum() #grouped sum of sales at Gender level
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('Gender')
ax1.set_ylabel('mathscore')
ax1.set_title("GENDER WISE MATHS SCORE")
var.plot(kind='bar')
```

```python
var = df.groupby('test preparation course').mathscore.sum()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('mathscore')
ax1.set_ylabel('maths score')
ax1.set_title("TEST PREPARATION VS MATHS SCORE")
var.plot(kind='line')
```
